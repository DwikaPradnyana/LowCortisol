### Cleaning Data

```python
# hapus baris duplikat
df = df.drop_duplicates()
print(f"Jumlah baris duplikat: {df.duplicated().sum()}")
```

```python
# standardisasi Employee_ID
# Ubah semua ke huruf KAPITAL & strip spasi; "nan" string -> NaN asli
df["Employee_ID"] = (
    df["Employee_ID"].astype(str).str.strip().str.upper().replace("NAN", np.nan)
)
print(df["Employee_ID"])
```

```python
# standarisasi Gender
gender_map = {
    "male": "Male",
    "female": "Female",
    "non-binary": "Non-binary",
    "prefer not to say": "Prefer not to say",
}
df["Gender"] = df["Gender"].str.strip().str.lower().map(gender_map)
print(df["Gender"])
```

```python
# standarisasi Work_Location
loc_map = {
    "remote": "Remote",
    "onsite": "Onsite",
    "hybrid": "Hybrid",
}
df["Work_Location"] = df["Work_Location"].str.strip().str.lower().map(loc_map)
print(df["Work_Location"])
```

```python
# standarisasi Stress_Level
stress_map = {
    "low": "Low",
    "medium": "Medium",
    "high": "High",
}
df["Stress_Level"] = df["Stress_Level"].str.strip().str.lower().map(stress_map)
print(df["Stress_Level"])
```

```python
# standarisasi Access_to_Mental_Health_Resources
mhr_map = {
    "yes": "Yes",
    "no": "No",
}
df["Access_to_Mental_Health_Resources"] = (
    df["Access_to_Mental_Health_Resources"].str.strip().str.lower().map(mhr_map)
)
print(df["Access_to_Mental_Health_Resources"])
```

```python
# standarisasi Productivity_Change
prod_map = {
    "increase": "Increase",
    "decrease": "Decrease",
    "no change": "No Change",
}
df["Productivity_Change"] = (
    df["Productivity_Change"].str.strip().str.lower().map(prod_map)
)
print(df["Productivity_Change"])
```

```python
# standarisasi Satisfaction_with_Remote_Work
sat_map = {
    "satisfied": "Satisfied",
    "unsatisfied": "Unsatisfied",
    "neutral": "Neutral",
}
df["Satisfaction_with_Remote_Work"] = (
    df["Satisfaction_with_Remote_Work"].str.strip().str.lower().map(sat_map)
)
print(df["Satisfaction_with_Remote_Work"])
```

```python
# standarisasi Physical_Activity
# NaN -> isi dengan modus
phys_map = {
    "daily": "Daily",
    "weekly": "Weekly",
}
df["Physical_Activity"] = df["Physical_Activity"].str.strip().str.lower().map(phys_map)
mode_physical = df["Physical_Activity"].mode()[0]
df["Physical_Activity"] = df["Physical_Activity"].fillna(mode_physical)
print(f"Modus Physical_Activity: {mode_physical}")
print(df["Physical_Activity"])
```

```python
# standarisasi Sleep_Quality
sleep_map = {
    "good": "Good",
    "average": "Average",
    "poor": "Poor",
}
df["Sleep_Quality"] = df["Sleep_Quality"].str.strip().str.lower().map(sleep_map)
print(df["Sleep_Quality"])
```

```python
# standarisasi Mental_Health_Condition
# NaN berarti tidak ada kondisi yang teridentifikasi -> isi dengan 'None'
df["Mental_Health_Condition"] = df["Mental_Health_Condition"].fillna("None")
print(df["Mental_Health_Condition"])
```

```python
# standarisasi kolom numerik dan menghapus outlier
num_cols = [
    "Age",
    "Years_of_Experience",
    "Hours_Worked_Per_Week",
    "Number_of_Virtual_Meetings",
    "Work_Life_Balance_Rating",
    "Social_Isolation_Rating",
    "Company_Support_for_Remote_Work",
]
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Batas nilai wajar berdasarkan domain knowledge
bounds = {
    "Age": (18, 70),
    "Years_of_Experience": (0, 50),
    "Hours_Worked_Per_Week": (0, 80),
    "Number_of_Virtual_Meetings": (0, 30),
    "Work_Life_Balance_Rating": (1, 5),
    "Social_Isolation_Rating": (1, 5),
    "Company_Support_for_Remote_Work": (1, 5),
}

total_outliers = 0
for col, (lo, hi) in bounds.items():
    mask = (df[col] < lo) | (df[col] > hi)
    total_outliers += mask.sum()
    df[col] = df[col].where(~mask)  # nilai di luar batas -> NaN
```

```python
df.to_csv("dataKaggle_clean.csv", index=False)
print(f"Data bersih tersimpan. Shape akhir: {df.shape}")
```
